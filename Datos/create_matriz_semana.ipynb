```python
import pandas as pd
```

```python
df = pd.read_csv('./files/mq7_sensor01_lunes.csv')
df
```

```python
# Asegúrate de que la columna 'Fecha' esté en formato datetime
df_filtrado = df[df['Fecha'] == '2023-11-06']
df_filtrado['Fecha_Hora'] = pd.to_datetime(df_filtrado['Fecha'] + ' ' + df_filtrado['Hora'])

# Agrupa por hora y calcula la media de los valores
df_agrupado = df_filtrado.groupby(pd.Grouper(key='Fecha_Hora', freq='H')).mean()

list_data = df_agrupado['valor'].agg(list)
# Imprime el DataFrame agrupado
print(list_data)
```
